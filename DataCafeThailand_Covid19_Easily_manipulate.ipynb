```python
import pandas as pd
```

```python
!ls ./time_series_19-covid-Confirmed.csv
```

```python
conf = pd.read_csv('./time_series_19-covid-Confirmed.csv')
conf.tail()
```

```python
deaths = pd.read_csv('./time_series_19-covid-Deaths.csv')
deaths.tail()
```

```python
recov = pd.read_csv('./time_series_19-covid-Recovered.csv')
recov.tail()
```

```python
conf['file'] = 'Confirmed'
conf.head()
```

```python
deaths['file'] = 'Deaths'
deaths.head()
```

```python
recov['file'] = 'Recovered'
recov.head()
```

```python
all_data = pd.concat([conf,deaths,recov], axis = 0)
all_data
```

```python
all_data.columns
```

```python
country_lv = all_data.drop(columns=['Province/State', 'Lat', 'Long'])
country_lv
```

```python
country_lv.columns
```

```python
country_lv = country_lv.groupby(by = ['Country/Region','file']).sum().reset_index()
country_lv
```

```python
select_col_date = [col for col in country_lv.columns if col not in ('Country/Region','file')] 
select_col_date
```

```python
country_lv_melt = country_lv.melt(id_vars=['Country/Region','file'], value_vars=select_col_date)
country_lv_melt
```

```python
country_lv_melt.rename(columns={
    'file':'Type',
    'variable':'Date',
    'value':'Case'
}, inplace=True)
country_lv_melt
```

```python
last_date = select_col_date[-1].replace('/','.')
last_date
```

```python
country_lv_melt.to_csv(f'./covid19_{last_date}.csv',index=False)
```
